```python
import os
import boto3
from botocore.exceptions import NoCredentialsError
from dotenv import load_dotenv
```

```python
load_dotenv()
aws_access_key=os.getenv("aws_access_key_id")
aws_secret_access=os.getenv("aws_secret_access_key")
region=os.getenv("region_name")
bucket_name=os.getenv("bucket_name")
local_dir = "C:\\Users\\jayak\\Downloads\\Python workspace\\infosys final work\\pythonProject\\ty3"
```

```python
client=boto3.client('s3',aws_access_key_id=aws_access_key,aws_secret_access_key=aws_secret_access,region_name=region)
```

```python
def upload_file_to_s3(local_file_path, bucket_name, s3_file_path):
    try:
        client.upload_file(local_file_path, bucket_name, s3_file_path)
        print(f"Upload Successful: {local_file_path} to s3://{bucket_name}/{s3_file_path}")
    except FileNotFoundError:
        print(f"The file was not found: {local_file_path}")
    except NoCredentialsError:
        print("Credentials not available")
```

```python
for root, dirs, files in os.walk(local_dir):
    for file in files:
        local_file_path = os.path.join(root, file)
        relative_path = os.path.relpath(local_file_path, local_dir)
        s3_file_path = relative_path.replace("\\", "/")  # Replace backslashes with forward slashes for S3
        s3_file_path = "Jayakrishna/" + s3_file_path
        upload_file_to_s3(local_file_path, bucket_name, s3_file_path)
```
